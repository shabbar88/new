```python
# write a program to find the area of different types of polygon
class Polygon:
    def sq_area(Self,side):
        Self.side=side
        return side*side
    
    def rect_area(Self,length,width):
        Self.length=length
        Self.width=width
        return length*width
    
    def tri_area(Self,side1,side2,side3):
        Self.side1=side1
        Self.side2=side2
        Self.side3=side3
        semi_peri=(side1+side2+side3)/2
        a1=semi_peri-side1
        a2=semi_peri-side2
        a3=semi_peri-side3
        a4=a1*a2*a3
        ans=a4**(1/2)
        return ans
    
    def penta_area(Self,side,apothem):
        Self.side=side
        Self.apothem=apothem
        return (5/2)*side*apothem
    
    def hexa_area(Self,side):
        Self.side=side
        return (3/2)*1.73*side*side
    
    def hepta_area(Self,side):
        Self.side=side
        return 3.634*side*side
    
    def octa_area(Self,side):
        Self.side=side
        return 2*(side)*(side)*(2.414)
    
class calcArea():
    p=Polygon()
    x=input("Enter the shape of polygon u wanna calculate area..[triangle,square,rectangle,pentagon,hexagon,heptagon,octagon]..")
    
    if x=='triangle':
            side1 = int(input("Enter side1 "))
            side2=int(input("Enter side2 "))
            side3 = int(input("enter side3 "))
            print(f"Area of triangle is: {p.tri_area(side1,side2,side3)}")
            
    if x=='square':
            side = int(input("Enter side "))
            print(f"Area of square is :{p.sq_area(side)}")
                
                
    if x=='rectangle':
            side1 = int(input("Enter side1 "))
            side2=int(input("Enter side2"))
            print(f"Area of rectangle is :{p.rect_area(side1,side2)}")
            
            
    if x=='pentagon':
            side = int(input("Enter side "))
            apoth=int(input("Enter apothem"))
            print(f"Area of pentagon : {p.penta_area(side,apoth)}")
            
    if x=='heptagon':
            side = int(input("Enter side "))
            print(f"Area of heptagon : {p.hepta_area(side)}")
            
    if x=='hexagon':
            side = int(input("Enter side "))
            print(f"Area of hexagon :{p.hexa_area(side)}")
            
    if x=='octagon':
            side = int(input("Enter side "))
            print(f"Area of octagon :{p.octa_area(side)}")
    
area=calcArea()
    
            
            
        
        
        
                
        
        

           
        
       
        
        
    
    
        
        
    
        
       
        
```
